```python
for i in range(1,6): #outer loop
   for j in range(1,11): #inner loop
    print(f"{i} * {j} = {i*j}")

   print()
```

```python
for num in range(1,11):
  factorial = 1
  for i in range(1,num + 1):
    factorial *=i
  print(f"The factorial of {num} is {factorial}")
  print()
```
